```python
import pandas as pd

df = pd.read_csv("../data/All_Pokemon.csv")
df
```

```python
df.columns
```

```python
df.info()
```

```python
df.columns = df.columns.str.replace(" ", "_").str.lower()
df.columns
```

```python
df.isnull().sum()
```

```python
df["type_2"].fillna("None", inplace=True)
df.isnull().sum()
```

```python
df["bmi"].describe()
```

```python
outliers = df[(df["bmi"] == 0) | (df["bmi"] > 1000)]
outliers
```

```python
df["bmi"] = df["bmi"].apply(lambda x: min(x, 1000))
df["bmi"].describe()
```

```python
df["type_1"].unique()
df["type_2"] = df["type_2"].str.replace(" ", "_").str.lower()
df["type_1"] = df["type_1"].str.replace(" ", "_").str.lower()
df["name"] = df["name"].str.replace(" ", "_").str.lower()
df["abilities"] = df["abilities"].str.replace("_", " ").str.lower()
df
```

```python
count_bmi_less_than_1 = df[df["bmi"] <= 1].shape[0]
print(f"Number of Pokémon with BMI less than 1: {count_bmi_less_than_1}")
```

```python
df["bmi_clipped"] = df["bmi"].clip(lower=1, upper=100)

print(df["bmi_clipped"].describe())
```

```python
df["atk_def_ratio"] = df["att"] / df["def"]


df["offensive_power"] = df["att"] + df["spa"]

df["defensive_power"] = df["def"] + df["spd"]

df[["atk_def_ratio", "offensive_power", "defensive_power"]].describe()
```
