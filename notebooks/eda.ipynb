```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/All_Pokemon_cleaned.csv")

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x="type_1", order=df["type_1"].value_counts().index)
plt.title("Distribution of Primary Pokmon Types")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="legendary")
plt.title("Legendary vs Non-Legendary Pokmon")
plt.xticks([0, 1], ["Non-Legendary", "Legendary"])
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 8))
stats_cols = ["hp", "att", "def", "spa", "spd", "spe", "bst", "mean"]
sns.heatmap(df[stats_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Pokmon Stats")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="generation", y="bst")
plt.title("Base Stat Total by Generation")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="att", y="def", hue="type_1", alpha=0.7)
plt.title("Attack vs Defense by Primary Type")
plt.tight_layout()
plt.show()
```

```python
key_stats = ["att", "def", "spa", "spd", "spe", "hp"]
sample_types = df["type_1"].value_counts().index[:6]
df_pair = df[df["type_1"].isin(sample_types)]

sns.pairplot(df_pair, vars=key_stats, hue="type_1", palette="Set2")
plt.suptitle("Pairplot of Key Stats by Type 1", y=1.02)
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(
    x="type_1",
    y="atk_def_ratio",
    data=df[df["type_1"].isin(sample_types)],
    palette="Set2",
)
plt.title("Attack-Defense Ratio by Type 1")
plt.xticks(rotation=45)
plt.show()


plt.figure(figsize=(8, 5))
sns.histplot(df["bmi"], kde=True, color="skyblue")
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.show()

plt.figure(figsize=(8, 5))
sns.kdeplot(
    data=df,
    x="height",
    hue="type_1",
    common_norm=False,
    palette="Set2",
    alpha=0.5,
    legend=True,
)
plt.title("Height Distribution by Type 1")
plt.show()

plt.figure(figsize=(8, 5))
sns.kdeplot(
    data=df,
    x="weight",
    hue="type_1",
    common_norm=False,
    palette="Set2",
    alpha=0.5,
    legend=True,
)
plt.title("Weight Distribution by Type 1")
plt.show()
```
